# Example 1

Print the prime numbers from 1 to 100.

```python
# Example 1
for i in range(101):
  prime=True
  j=2
  while prime and j<i:
    if (i % (j))==0:
      prime=False
    j+=1
  if prime:
    print(i)
```

```python
def prime(x):
  for i in range(2,x):
    prime = True
    j =2
    while prime and j<i:
      if(i%j)==0:
        prime = False
      j+=1
    if prime:
      print(i)
prime(10)
```

# Task 2

Modify the previous task and define class finding the prime numbers in selected range.


```python
class Prime:
  def __init__(self,limit):
    self.limit = limit

  def is_prime(self,num):
    if num < 2:
      return False
    for i in range(2, int(num ** 0.5)+ 1):
      if num % i ==0:
        return False
    return True

  def primes(self):
    for i in range(2,self.limit):
      if self.is_prime(i):
        print(i)

p = Prime(10)
p.primes()
```

# Task 9

The sigmoid function is frequently used activation of a neuron. It is defined by

$$y=\sigma(x)=\frac{1}{1+e^{-x}}$$

Draw a graf of this function for $x\in <-8,8>$.

```python
def sgm(x):
  return 1/(1+np.exp(-x))

x = np.linspace(-8,8,100)
print(x)

plt.axhline(0, color = 'gray', linestyle = ':')
plt.axhline(0.5, color = 'gray', linestyle = ':')
plt.axhline(1, color = 'gray', linestyle = ':')
plt.axvline(color = 'gray')

plt.plot(x, sgm(x), label = 'sigmoid')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Task 10

The RELU function is frequently used activation function of neurons in deep neural networks. It is defined by

$$y=\mbox{relu}(x)=\mbox{max}(0, x)$$

Draw a graf of this function for $x\in <-10,10>$.

```python
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
  return np.maximum(0, x)

x = np.linspace(-10,10,100)
print(x)

plt.plot(x, relu(x), label='relu')
plt.axhline(0, color='gray', linestyle=':')
plt.axvline(color='gray')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Task 11

The Gaussian function is frequently used as an activation function of RBF neurons. It is defined by the equation

$$y=g(x)=e^{\frac{-x^2}{2\sigma^2}}$$

Draw a graf of this function for $x\in <-10,10>$ and $\sigma=\{0.2,0.5,1,2,5\}$.

```python
def rbf(x,sigma):
  return np.exp((-x**2)/(2*sigma**2))

x = np.linspace(-10,10,100)
sigma = [0.2,0.5,1,2,5]
for sig in sigma:
  plt.plot(x, rbf(x,sig), label='sigma='+str(sig))
plt.axhline(y = 0 , color = 'gray', linestyle = ':')
plt.axvline(color = 'gray')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Task 12

Draw response of the two input perceptron neuron defined by

$$y=\sigma(\sum_{i=0}^{1}w_ix_i + b)$$

where $w_0=10$,$w_1=10$, $b=-10$.

```python
from mpl_toolkits.mplot3d import Axes3D
def per(x0,x1,w0 = 10,w1 = 10, b = -10):
  z = w0 * x0 + w1 * x1 + b
  return 1 / (1 + np.exp(-z))

x0_values = np.linspace(0, 1, 100)
x1_values = np.linspace(0, 1, 100)

X0, X1 = np.meshgrid(x0_values, x1_values)

Y = per(X0, X1)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X0, X1, Y, cmap='viridis', edgecolor='none')

# Labels and title
ax.set_xlabel('Input x0')
ax.set_ylabel('Input x1')
ax.set_zlabel('Output y')
ax.set_title('3D Response of the Perceptron Neuron')

# Show the plot
plt.show()

```

# Task 13

Draw response of the two input RBF neuron defined by

$$y=g(\sqrt{\sum_{i=0}^{1}(x_i-c_i)^2})$$

where $c_0=0.5$, $c_1=0.5$, $\sigma=0.15$.

```python
import math
def rb( x0, x1, c0 = 0.5, c1=0.5, sig = 0.15):
  a = np.sqrt((x0-c0)**2 + (x1-c1)**2)
  return np.exp((-a**2)/(2*sig**2))

x0_values = np.linspace(0, 1, 100)
x1_values = np.linspace(0, 1, 100)

X0, X1 = np.meshgrid(x0_values, x1_values)

Y = rb(X0, X1)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(X0,X1,Y, cmap = 'viridis', edgecolor = 'none')

ax.set_xlabel('Input x0')
ax.set_ylabel('Input x1')
ax.set_zlabel('Output y')
ax.set_title('3D Response of the RBF Neuron')

plt.show()
```
